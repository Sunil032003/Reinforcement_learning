```python
n=int(input("\n Enter the number of episodes (games:)"))
returns=[]
#input returns for each episodes
for i in range(n):
  value=float(input(f"Enter return for episodes {i+1}="))
  returns.append(value)

print("\nReturns obtained from state A")
print("--"*35)
total=0
for i in range(n):
  print(f"Episodes {i+1}: {returns[i]}")
  total+=returns[i]

#calculate mean

mean=total/n
print("--"*35)
print(f"Total returns={total}")
print(f"number of episodes={n}")
print(f"mean (Average)={mean:.2f}")

#monte carlo prediction

print("\nMonte carlo prediction")
print("--"*35)
print(f"Estimated of state A (v(A))={mean:.2f}")
print("\nExplanation")
print("monte carlo prediction estimates the value of state")
print("by taking the average return obtained from multiple episodes")
```
